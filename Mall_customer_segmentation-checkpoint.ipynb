```python
df['Age_log'] = np.log(df['Age'] + 1)
df['Income_log'] = np.log(df['Annual Income (k$)'] + 1)
```

```python
df=df.drop(columns=["Age", "Annual Income (k$)","CustomerID"])
```

```python
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
```

```python
df['Gender'].value_counts()
```

```python
gender_map = dict(zip(le.classes_, le.transform(le.classes_)))
print(gender_map)
```

```python
scaler = StandardScaler()
df_for_scale = df.drop(columns=["Gender"])
scaled_data = scaler.fit_transform(df_for_scale)

df_scaled = pd.DataFrame(scaled_data, columns=df_for_scale.columns)
df_scaled["Gender"] = df["Gender"].values 
```

```python
for n in [2, 3, 4]:
    pca = PCA(n_components=n, random_state=42)
    X_pca = pca.fit_transform(df)
    total_variance = np.sum(pca.explained_variance_ratio_)
    print(f"{n} components → Total explained variance: {total_variance:.2f}")
    
    col_names = [f"PCA{i+1}" for i in range(n)]
    df_pca = pd.DataFrame(X_pca, columns=col_names)

    explained_variance = pca.explained_variance_ratio_
    total_var = np.sum(explained_variance)

    print(f"\n{n} Components:")
    print(f"Explained variance by each: {explained_variance}")
    print(f"Total explained variance: {total_var:.2f}")
```

```python
pca = PCA(n_components=3, random_state=42)
X_pca = pca.fit_transform(df)  

df_pca = pd.DataFrame(X_pca, columns=[f"PCA{i+1}" for i in range(3)])
```

```python
models = {
    "KMeans": KMeans(n_clusters=3, random_state=42),  
    "Hierarchical": AgglomerativeClustering(n_clusters=3),
    "DBSCAN": DBSCAN(eps=0.5, min_samples=5)          
}
```

```python
results = []

for name, model in models.items():
    labels = model.fit_predict(df_pca)
    
    if len(set(labels)) > 1:
        score = silhouette_score(df_pca, labels)
    else:
        score = -1
    results.append((name, score))

df_results = pd.DataFrame(results, columns=["Model", "Silhouette Score"])
print(df_results.sort_values(by="Silhouette Score", ascending=False))
```

```python
best_model = KMeans(n_clusters=3, random_state=42)
best_labels=best_model.fit_predict(df_pca)
```

```python
df_final = df.copy()
df_final["Cluster"] = best_labels
df_final.head()
```
